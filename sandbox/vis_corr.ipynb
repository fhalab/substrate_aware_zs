```python
matrix = corr_df.values
```

```python
# Define slices for each region
regions = {
    "top_left": matrix[:11, :11],          # Top-left 11x11
    "top_middle1": matrix[:11, 11:12],    # First top-middle 11x1
    "top_middle2": matrix[:11, 12:13],    # Second top-middle 11x1
    "top_right": matrix[:11, 13:],        # Top-right 11x9
    "row1_left": matrix[11:12, :11],      # First middle row 1x11
    "row1_middle1": matrix[11:12, 11:12], # First middle row 1x1
    "row1_middle2": matrix[11:12, 12:13], # Second middle row 1x1
    "row1_right": matrix[11:12, 13:],     # First middle row 1x9
    "row2_left": matrix[12:13, :11],      # Second middle row 1x11
    "row2_middle1": matrix[12:13, 11:12], # Second middle row 1x1
    "row2_middle2": matrix[12:13, 12:13], # Second middle row 1x1
    "row2_right": matrix[12:13, 13:],     # Second middle row 1x9
    "bottom_left": matrix[13:, :11],      # Bottom-left 9x11
    "bottom_middle1": matrix[13:, 11:12], # Bottom-middle 9x1
    "bottom_middle2": matrix[13:, 12:13], # Bottom-middle 9x1
    "bottom_right": matrix[13:, 13:]      # Bottom-right 9x9
}

# Compute the mean of each region
region_means = {key: np.mean(value) for key, value in regions.items()}

# Construct the 4x4 matrix from region means
aggregated_matrix = np.array([
    [region_means["top_left"], region_means["top_middle1"], region_means["top_middle2"], region_means["top_right"]],
    [region_means["row1_left"], region_means["row1_middle1"], region_means["row1_middle2"], region_means["row1_right"]],
    [region_means["row2_left"], region_means["row2_middle1"], region_means["row2_middle2"], region_means["row2_right"]],
    [region_means["bottom_left"], region_means["bottom_middle1"], region_means["bottom_middle2"], region_means["bottom_right"]],
])

# Display the resulting 4x4 matrix
print("Aggregated 4x4 Matrix:")
print(aggregated_matrix)
```

```python
corr_df["ParLQ-avg"] = corr_df[[c for c in lib_order if "ParLQ" in c]].mean(axis=1)
corr_df["PfTrpB-avg"] = corr_df[[c for c in lib_order if "PfTrpB" in c]].mean(axis=1)

groups = {"ParLQ-avg": [c for c in lib_order if "ParLQ" in c], "PfTrpB-avg": [c for c in lib_order if "PfTrpB" in c]}

for new_row, rows in groups.items():
    corr_df.loc[new_row] = corr_df.loc[rows].mean()
corr_df
```
